```python
bins= [19,25,31,36,40]
labels = ['19-24','25-30','31-35','36-40']
dataset['AgeGroup'] = pd.cut(dataset['Age'], bins=bins,labels=labels, right=False)
```

```python
def basic_statistics(df):
    print("\nIris-setosa")
    print(df[df.species=='setosa'].describe().transpose())
    print("\nIris-versicolor")
    print(df[df.species=='versicolor'].describe().transpose())
    print("\nIris-virginica")
    print(df[df.species=='virginica'].describe().transpose())
basic_statistics(iris_dataset)
```
