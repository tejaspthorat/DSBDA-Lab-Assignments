```python
import pandas as pd
```

```python
df = pd.read_csv("/content/HousingData.csv")
```

```python
df.describe()
```

Checking for null values

```python
df.isnull().sum()
```

Replacing null values with mean

```python
df.fillna(df.mean(), inplace=True)
df.isnull().sum()
```

Standard scaling the values

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_values = scaler.fit_transform(df)

scaled_df = pd.DataFrame(scaled_values, columns=df.columns)

scaled_df
```

Handling Outliers

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.title('Dot Plot for Outliers Detection')
plt.boxplot(scaled_df)
plt.xlabel('Features')
plt.ylabel('Values')
plt.show()

Q1 = scaled_df.quantile(0.25)
Q3 = scaled_df.quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

for column in scaled_df.columns:
  scaled_df[column] = scaled_df[column].clip(lower=lower_bound[column], upper=upper_bound[column])

plt.figure(figsize=(10, 6))
plt.title('Dot Plot After Outliers Handling')
plt.boxplot(scaled_df)
plt.xlabel('Features')
plt.ylabel('Values')
plt.show()
```

Splitting the dataset into train and test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(scaled_df, df['MEDV'], test_size=0.2, random_state=1)
```

Linear Regression using .fit() from sklearn

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
r2 = r2_score(y_test, y_pred)

print("Mean squared error:", mean_squared_error(y_test, y_pred))
print("R-squared score:", r2)
```
