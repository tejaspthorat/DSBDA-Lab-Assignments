```python
import seaborn as sns

iris_dataset = sns.load_dataset("iris")
```

```python
import matplotlib.pyplot as plt

for column in iris_dataset.columns:
    plt.figure(figsize=(5, 3))
    plt.hist(x=iris_dataset[column], bins=5, edgecolor="black")
    plt.title(f"Distribution of {column}")
    plt.xlabel(column)
    plt.ylabel("Frequency")
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

for column in iris_dataset.select_dtypes(include=["int", "float"]):
    plt.figure(figsize=(5, 3))
    sns.boxplot(x=iris_dataset[column])
    plt.title(f"Box Plot of {column}")
    plt.xlabel(column)
    plt.ylabel("Value")
    plt.show()
```
