```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
df
```

```python
df.isnull().sum()
```

```python
df['deck'].value_counts()
```

```python
df['deck']=df['deck'].ffill()
```

```python
df['deck'][0]='C'
```

```python
df.isnull().sum()
```

```python
df=df.drop('embark_town',axis=1)
```

```python
df.isnull().sum()
```

```python
df['age']=df['age'].fillna(df['age'].mean())
```

```python
df['embarked'].value_counts()
```

```python
df['embarked']=df['embarked'].fillna('S')
```

```python
df.isnull().sum()
```

```python
sns.catplot(x ="who", hue ="survived", kind ="count", data = df)
```

```python
ax=sns.scatterplot(data=df,x='class',y='age',hue='survived')
sns.move_legend(ax,  "upper right")
```

```python
ax=sns.scatterplot(data=df,x='age',y='fare',hue='survived')
sns.move_legend(ax,"upper right")
```

```python
sns.violinplot(x ="sex", y ="age", hue ="survived", data = df, split = True,palette=['pink','black'])
```

```python
bins=[0,100,200,300,400,500]
sns.histplot(data=df,x='fare',bins=bins)
```
